```python
def target_encoding(data: pd.DataFrame, filling_value: str, target_value: str = 'Price_euros') -> pd.DataFrame:
    grouped = data.groupby(filling_value)[target_value].mean()
    temp = data.copy(deep=True)
    for item in temp[filling_value].unique():
        temp.loc[temp[filling_value] ==  item, filling_value] = grouped[item]
    return temp        
```

### ÐžÐ±Ñ€Ð°Ð±Ð¾Ñ‚ÐºÐ° Company

```python
data = target_encoding(data, 'Company')
```

### ÐžÐ±Ñ€Ð°Ð±Ð¾Ñ‚ÐºÐ° TypeName
